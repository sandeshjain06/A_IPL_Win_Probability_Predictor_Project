```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
match=pd.read_csv("D:\Sandesh\Data Science\Project - 1\Campusx_Project\matches.csv")
delivery=pd.read_csv("D:\Sandesh\Data Science\Project - 1\Campusx_Project\deliveries.csv")
match.head()
```

```python
match.shape
```

```python
delivery.shape
```

```python

# 1. Calculate the total runs scored  in innings 1.
total_runs= delivery[delivery['inning'] == 1].groupby(['match_id','inning'])['total_runs'].sum().reset_index()
total_runs
```

```python
match = match.merge(total_runs,left_on='id',right_on='match_id')
match.head()
```

```python
match['team1'].unique()
```

```python

# 2.Rename the team names who are only currently playing
# Data of top 8 teams 

match['team1']=match['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match['team2']=match['team2'].str.replace('Delhi Daredevils','Delhi Capitals')
match['team1']=match['team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match['team2']=match['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
```

```python
# Remove the unwanted team names
teams = ['Sunrisers Hyderabad', 'Mumbai Indians','Royal Challengers Bangalore',
       'Kolkata Knight Riders', 'Delhi Capitals', 'Kings XI Punjab',
       'Chennai Super Kings', 'Rajasthan Royals' ]

match=match[match['team1'].isin(teams)]
match=match[match['team2'].isin(teams)]
match.head()
```

```python
match.shape
```

```python
# 3. remove the columns where dl_applied = 1

match = match[match['dl_applied'] != 1]
match.shape
```

```python
match = match[['match_id','city','winner','total_runs']]
delivery_df = match.merge(delivery,on='match_id')
```

```python
delivery_df.head()
```

```python
# 4. Select only inning where innings=2 , bec we will work on 
# second innings only

delivery_df = delivery_df[delivery_df['inning'] == 2]
delivery_df.shape
```

```python
delivery_df.head()
```

```python

# check the runs scored every ball and do the cumsum to add the scores.
# and deduct it from the total score to get the current score required.

delivery_df['current_score']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
delivery_df.shape
```

```python


delivery_df['runs_left']= delivery_df['total_runs_x'] - delivery_df['current_score']
delivery_df.shape
```

```python
delivery_df.head()
```

```python

# to get the overs remaining

delivery_df['balls_left']=126-(delivery_df['over']* 6 + delivery_df['ball'])
delivery_df.head()
```

```python

# Now , get the No of Player dismissed 
# we can get that by player-dismissed feature 

delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna(0)
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x : x if x==0 else "1")
delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype('int')

wickets=delivery_df.groupby('match_id')['player_dismissed'].cumsum()
delivery_df['wickets']= 10 - wickets
```

```python
delivery_df.head()
```

```python

# now Calculate the current run-rate
# crr = total runs  scored / total overs played

delivery_df['crr']=(delivery_df['current_score'] * 6) / (120-delivery_df['balls_left'])

# rrr 

delivery_df['rrr']=(delivery_df['runs_left'] * 6) / (delivery_df['balls_left'])
delivery_df.head()
```

```python

# Calculate the result


def result(row):
    return 1 if row['winner'] == row['batting_team'] else 0

delivery_df['result']=delivery_df.apply(result,axis=1)
```

```python
delivery_df.head()
```

```python
# Extract only the required fields

final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','crr','rrr','result']]
final_df.head()
```

```python
final_df.shape
```

```python
# reshuffle the dataset

final_df = final_df.sample(final_df.shape[0])
final_df
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] !=0]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score,accuracy_score
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.metrics import Ran
```

```python
x=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
trf = ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team',
                                                    'city'])],
    remainder='passthrough')
```

```python
pipe=Pipeline(
    steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear'))])

pipe.fit(x_train,y_train)
```

```python
y_predict=pipe.predict(x_test)
```

```python
print(accuracy_score(y_test,y_predict))
```

```python
import pickle

pickle.dump(pipe,open('IplScore.pkl','wb'))
```
